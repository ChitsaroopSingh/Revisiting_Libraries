Making 1 dimensional arrays using numpy

```python
import numpy as np
```

numpy.arange creates arrays with regularly incrementing values

```python
x=np.arange(1,10) #creates an array from 1 to 9. excludes the 'stop'
print(x)
print(type(x)) #<class 'numpy.ndarray'>
print(x[3]) #element at index 3 (=4)
```

```python
a=np.arange(0,5,0.5) #increment of 0.5
print(a)
```

numpy.linspace creates arrays with a specified number of elements spaced equally between the specified beginning and end values including 'stop'

```python
x=np.linspace(1,10,5) #5 elements spaced equally
print(x)
```
