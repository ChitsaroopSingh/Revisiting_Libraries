Creating 2-D arrays

```python
import numpy as np
```

np.eye(n,m) defines a 2-D identity matrix

```python
a=np.eye(3) #identity matrix of 3x3
print(a)
print("\n")
b=np.eye(2,4) #2 diagonal elements have value 1
print(b)
```

np.diag creates a diagonal matrix

```python
a=np.diag([1,2,3])
print(a)
print("\n")

b=np.diag([1, 2, 3], 2) #the second argument adds a the number of new rows and columns (from the bottom left corner) with the element zero
print(b)
print("\n")

c=np.array([[1, 2], [3, 4]])
print(c)
```

numpy.zeros will create an array filled with 0 values with the specified shape

```python
print(np.zeros((2,3)))
print(np.zeros((2,3,2)))
```

numpy.ones will create an array filled with 1 values

```python
print(np.ones((2,3)))
```
