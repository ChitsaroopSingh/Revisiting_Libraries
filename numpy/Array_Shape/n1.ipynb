```python
import numpy as np
```

.shape tell the number of rows and columns it has

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr.shape)

x=np.arange(10)
print(x)
x.shape=(2,5) #x is now 2 dimensional
print(x)
print(x[1,3])
```

numpy.reshape gives a new shape to an array without changing its data

```python
a = np.arange(6).reshape((3, 2))
print(a)
```
